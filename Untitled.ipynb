```python
import requests
from bs4 import BeautifulSoup
```

```python
page_cou=1
while True:
    url=f"https://quotes.toscrape.com/page/{page_cou}/"
    res=requests.get(url)
    soup=BeautifulSoup(res.text,"lxml")
    if not soup.select("div.quote"):
        break
    with open(f"scraped_data/quotes{page_cou}.html","w") as f:
        f.write(res.text)
    page_cou+=1

    
```

```python
page_cou
```

```python
ans=[]
for i in range(1,page_cou):
    with open(f"scraped_data/quotes{i}.html","r") as f:
        z=f.read()
        soup=BeautifulSoup(z,"lxml")
        all_quotes=soup.select("div.quote")
        x=[]
    
        for z in all_quotes:
            m=[]
            for tag in z.select(".tags .tag"):
                m.append(tag.get_text())
            if "life" in m:
                text=z.select_one("span.text").get_text()
                author=z.select_one("small.author").get_text()
                ans.append([text,author])


```
